```python
df = pd.read_csv('table_1.csv', encoding='utf-8')

def reverse_text(text):
    if isinstance(text, str):
        return text[::-1]
    return text

# Apply the function to the entire DataFrame
df_corrected = df.applymap(reverse_text)


df_corrected.to_csv('table_1.csv')
```

```python
file_path = "/content/drive/MyDrive/shared/Ali and Navid Tasks/PDFs/Copy of SelTajrobi91.pdf"

all_pages_df = pd.DataFrame(columns=['university',  'Programme-Code','Programme-Major-Name', 'Sub-Major-Name',
                           'Capacity-first', 'Capacity-Second', 'Female-Cap', 'Male-Cap', 'Notes'])
for pp in range(24, 49):
  content = tabula.read_pdf(file_path, lattice = False, guess = False, columns = [210, 235, 255, 275, 310, 400, 530,580],
                                  pandas_options={'header': None}, pages = str(pp))
  content = content[0]
  page_list = []
  the_university = ''
  for index, row in content.iterrows():
    if type (row[7]) == str and row[7].isnumeric():
        row_list = list(row.values)
        row_list += [the_university]
        row_list = row_list[::-1]
        page_list += [row_list]
    else:
      temp_university = ''
      for i in range(7, -1, -1):
        if type(row[i]) == str:
          temp_university += row[i]
      if 'دانشگاه' in temp_university or 'آموزشكده' in temp_university or 'دانشكده' in temp_university or 'مجتمع' in temp_university or 'مركز' in temp_university:
        the_university = temp_university
  page_df = pd.DataFrame(page_list, columns=['university',  'Programme-Code','Programme-Major-Name', 'Sub-Major-Name',
                           'Capacity-first', 'Capacity-Second', 'Female-Cap', 'Male-Cap',  'Notes'])
  #all_pages_df = all_pages_df.append(page_df, ignore_index=True)
  all_pages_df = pd.concat([all_pages_df,page_df], ignore_index=True)

all_pages_df.to_excel('SelTajrobi91.xlsx')
```
